```python
dash = df.groupby('client_id').count().reset_index()[['client_id','date_order']]
dash
```

```python
max_total_order_count = dash.quantile(0.95)
max_total_order_count
```

```python
dash = dash.loc[dash['date_order'] < float(max_total_order_count)].loc[dash['date_order'] > 2]
dash
```

```python
df = df.loc[df.client_id.isin(dash.client_id)].reset_index(drop=True)
df
```
